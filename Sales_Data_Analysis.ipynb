```python
data.drop(columns='Postal Code',inplace=True)
```

```python
data.isnull().sum()
```

```python
total_sales = data['Sales'].sum()
print(total_sales)
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data.set_index('Order Date', inplace=True)
```

```python
monthly_sales = data['Sales'].resample('M').sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
monthly_sales.plot()
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
best_selling_products = data.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
best_selling_products.plot(kind='bar')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.show()
```

```python
sales_by_category = data.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sales_by_category.plot(kind='bar')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
sales_by_region = data.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sales_by_region.plot(kind='bar')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```
